# Visualizing the quarterly - ESTransformer

```python
from google.colab import drive
drive.mount('/content/gdrive/')

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

path = 'gdrive/My Drive/Colab Notebooks/hybcast_epoch/'


#-------------------------------------------------------------------------------------------
# Extract series
windows_y_hat = np.load('/content/gdrive/My Drive/Colab Notebooks/hybcast_epoch/windows_y_hat_quarterly.npy')
windows_y = np.load('/content/gdrive/My Drive/Colab Notebooks/hybcast_epoch/windows_y_quarterly.npy')
y = np.load('/content/gdrive/My Drive/Colab Notebooks/hybcast_epoch/y_quarterly.npy')

train = pd.read_csv(path+'train_y.csv')
test = pd.read_csv(path+'test_y.csv')
predicted = pd.read_csv(path+'y_predicted.csv')
#-------------------------------------------------------------------------------------------
```

```python
#-------------------------------------------------------------------------------------------
# First series
id_str = 'Q23183'
id_num = 0
train_date = train.loc[(train['unique_id'] == id_str) & (train.index >= 1344397)]['ds']
test_y = test.loc[(test['unique_id'] == id_str)]
test_y_value =  test.loc[(test['unique_id'] == id_str)]['y']
test_date = test.loc[(test['unique_id'] == id_str)]['ds']

y_predicted = predicted.loc[(test['unique_id'] == id_str)]['y_hat']

final_y_hat = []
for i in range(len(windows_y_hat)):
  final_y_hat[i:i+len(windows_y_hat[i][id_num])] = windows_y_hat[i][id_num]

final_y_hat += y_predicted.tolist()
truth_y = y[id_num].tolist() + test_y_value.tolist()


plt.figure(figsize=(15,7))
plt.plot(truth_y, color='green',label='Sample')
plt.plot(np.arange(4, len(truth_y), 1), final_y_hat, color='blue',label='In-sample forecast')
plt.plot([30, 31, 32, 33, 34, 35, 36, 37], y_predicted, color='red', label='Out-sample forecast')

general_date = train_date.tolist() + test_date.tolist()

space = 8
plt.xticks(np.arange(0, len(general_date), space), labels=[general_date[i] for i in np.arange(0, len(general_date), space)])

plt.title(id_str, fontsize='x-large', fontweight='bold')

plt.grid(True)
plt.legend()
plt.show()
```
